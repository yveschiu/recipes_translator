```python
from pymongo import MongoClient

def _connect_mongo(host, port, username, password, db):
    """指定帳戶＆密碼和 MongoDB 建立連接"""
    if username and password:
        mongo_uri = "mongodb://%s:%s@%s/%s" % (username, password, host, db)
        conn = MongoClient(mongo_uri)
    else:
        conn = MongoClient(host, port)
    return conn[db]

def read_mongo(db, collection, query={}, host="localhost", port=27017, username=None, password=None):
    """從MongoDB讀取並存到 dataframe"""

    # Connect to MongoDB
    db = _connect_mongo(host=host, port=port, username=username, password=password, db=db)

    # Make a query to the specific DB and Collection
    cursor = db[collection].find(query)
    
    data = [ d for d in cursor]
    
    return data
```

```python
# 把全部資料從 MongoDB 拉回來
recipes_list = read_mongo(db="recipes", collection="recipes",
                          host="localhost", port=27017,
                          username=None, password=None)
```

```python
# 載入 google 翻譯模組
from googletrans import Translator
translator = Translator()
```

```python
def translate_ch_to_en(material_name):
    return translator.translate(material_name).text
```

```python
import time 
err_list = []
for recipe_num, recipe in enumerate(recipes_list[:100]):
    for material_num, material in enumerate(recipe["materials"]):
        ch = material.get("material_name")
        try:
            en = translate_ch_to_en(ch)
        except:
            err_msg = (recipe_num, material_num, ch)
            err_list.append(err_msg)
        
        material["material_name_en"] = en
#         time.sleep(0.05)
```
